```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
x = np.arange(1,11)
y = np.arange(1,11)
z = np.random.randn(10)

fig3,ax = plt.subplots(3,2)

ax[0,0].plot(x,y,label='first')
ax[0,0].set_title("first")
ax[0,0].set_xticks([1,3,4,5,7,9])
ax[0,0].yaxis.set_major_locator(ticker.MaxNLocator(n=10))


ax[0,1].plot(x,z,label='Second')
ax[0,1].set_title("Second")
ax[0,1].xaxis.set_major_locator(ticker.MultipleLocator(base=2))
ax[0,1].xaxis.set_minor_locator(ticker.FixedLocator([1,3,5,7,9]))
# ax[0,1].yaxis.set_major_locator(ticker.MaxNLocator(n=1))


ax[1,0].plot(y,z,label='first')
ax[1,0].set_title("first")


ax[1,1].plot(y,y,label='Second')
ax[1,1].set_title("Second")


z = np.random.randn(10)

ax[2,0].plot(x,y,label='first')
ax[2,0].set_title("first")

ax[2,1].plot(x,z,label='Second')
ax[2,1].set_title("Second")

plt.tight_layout()
plt.show()
```
