```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv")
```

```python
plt.figure(figsize=(12,6),num=1)
plt.scatter(df.petal_length[:50],df.petal_width[:50],color='r',label='Iris-Setosa')
plt.scatter(df.petal_length[50:100],df.petal_width[50:100],color='g',label='Iris-versicolor')
plt.scatter(df.petal_length[101:],df.petal_width[101:],color='b',label='Iris-virginica')
plt.title("Iris")
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(12,6),num=1)
plt.scatter(df.sepal_length[:50],df.sepal_width[:50],color='r',label='Iris-Setosa')
plt.scatter(df.sepal_length[50:100],df.sepal_width[50:100],color='g',label='Iris-versicolor')
plt.scatter(df.sepal_length[101:],df.sepal_width[101:],color='b',label='Iris-virginica')
plt.title("Iris")
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.legend(loc='best')
plt.show()
```

# Multidimensional Scatter Plot - 4D Scatter Plot

```python
plt.figure(figsize=(9,7),num=1)
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
z = 5 * np.random.randint(1,100,100)
color = np.random.randint(1,100,100)
plt.scatter(x,y,s=z,c=color,alpha=0.5,cmap='rainbow')
plt.title("4D plot")
plt.colorbar()
plt.show()
```
