```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
```

```python
plt.plot(x,y,marker='o')
plt.title("First Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.figure(figsize=(9,4),num=1)
plt.plot(x,y,marker='o',ls='--')
plt.title("Second Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.figure(figsize=(9,4),num=1)
plt.plot(x,y,marker='o',ls='--')
plt.title("Third Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.savefig("Third Plo")
```
