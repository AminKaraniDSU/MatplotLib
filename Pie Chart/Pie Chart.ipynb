# Pie Chart

```python
from matplotlib import pyplot as plt
```

```python
x = [22,18,13,10]
labels = ['Math','Physics','Chemistry','English']
plt.style.use('ggplot')
plt.pie(x,labels=labels,autopct='%1.2f%%')
plt.title("Marks")
plt.show()
```


# 2nd Chart

```python
plt.figure(figsize=(9,9),num=1)
x = [33.71,19.37,4.85,3.82,7.42,30.83]
labels = ['Samsung','Apple','Huawei','LG','Unknown','Others']
plt.pie(x,labels=labels,autopct='%1.2f%%',explode=[.2,0,0,0,0,0],shadow=True,startangle=90,counterclock=False,wedgeprops={"edgecolor":'k',"linewidth":'1',"linestyle":'--'})
plt.legend(loc = 2)
plt.show()
```
