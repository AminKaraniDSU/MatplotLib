# Bar Chart

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
years = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
sales= [18000,14000,14500,13000,17000,18000,20000,20500,30000,20750,20250]
```

```python
x_pos = np.arange(len(years))
```

```python
plt.bar(x_pos,sales,color='m',edgecolor='c',label="Sales",alpha=0.4)
plt.xticks(x_pos,years,rotation=45)
```

# Clustered bar chart

```python
years = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
sales_toyota= [18000,14000,14500,13000,17000,18000,20000,20500,30000,20750,20250]
sales_honda=  [12000,11000,16500,17000,12000,13000,21000,20000,15000,18750,20250]
sales_suzuki= [15000,16000,15500,12000,15000,20000,30000,18500,30000,17750,20050]
```

```python
width = 0.25
```

```python
x_pos = np.arange(len(years))
plt.figure(figsize=(12,6),num=1)
plt.bar(x_pos-width,sales_toyota,color='m',edgecolor='c',alpha=0.4,width=width, label='Toyota')
plt.bar(x_pos,sales_honda,color='c',edgecolor='c',alpha=0.4,width=width, label='Honda')
plt.bar(x_pos+width,sales_suzuki,color='r',edgecolor='c',alpha=0.4,width=width, label='Suzuki')
plt.xticks(x_pos,years,rotation=45)
plt.legend(loc=2)
```
