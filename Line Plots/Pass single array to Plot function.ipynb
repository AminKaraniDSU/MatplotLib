```python
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
```

```python
plt.figure(figsize=(9,4),num=1)
plt.plot(x,y,color='c',marker='o')
plt.plot(x,y**2,marker='o',color='m')
plt.plot(x,y**2.5,marker='o',color='r')
```

```python
plt.figure(figsize=(9,4),num=1)
plt.plot(x,y,x,y**2,x,y**2.5,marker='o')
```

```python
plt.figure(figsize=(9,4),num=1)
plt.plot(x,y,"b--d",label="X")
plt.plot(x,y**2,"g:o",label="X*2")
plt.plot(x,y**3,"r-*",label="X*3")
plt.legend(loc=9)
plt.show()
```

```python
plt.figure(figsize=(9,4),num=1)
plt.plot(x,y,"b--d")
plt.plot(x,y**2,"g:o")
plt.plot(x,y**3,"r-*")
plt.legend(['X','X*2','X*3'],loc='Best')
plt.show()
```
